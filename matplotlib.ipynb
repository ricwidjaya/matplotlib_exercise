### Figure means the graphic without axis, you can imagine it as a blank canvas for data.

We are doing it because we want the most flexibility when we doing plotting.
Below are the ways to do plotting with Object-Oriented API.

```python
plt.plot();
```

```python
# 1st method of plotting

# Create a figure without axes
fig = plt.figure()

# Add some axes
ax = fig.add_subplot()

plt.show()
```

```python
# 2nd method of plotting

# Create a figure without axes
fig = plt.figure()

# Add some axes
ax = fig.add_axes([1, 1, 1, 1]) # [left, bottom, width, hight]

# Add some data
ax.plot(x, y)

plt.show()
```

```python
# 3rd method of plotting (recommended)

# Save line of code

# subplots() -> without arguments returns a Figure and a single Axes.
# This is actually the simplest and recommended way of creating a single Figure and Axes.

fig, ax = plt.subplots()
ax.plot(x, y);
```

# Matplotlib example workflow 

```python
# 0. Import matplotlib and get it ready for Jupyter Notebook. 
#如果不想要每次在 jupyter notebook 顯示都寫 .show()，可以在 jupyter notebook 加上 %matplotlib inline

# %matplotlib inline
# import matplotlib as plt

# 1. Prepare data

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot

fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data

ax.plot(x, y)

# 4. Customize plot

ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (You save the whole figure)

fig.savefig("image/Simple_Plot.png")
```
